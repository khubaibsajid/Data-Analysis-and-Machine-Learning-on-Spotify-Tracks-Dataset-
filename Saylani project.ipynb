# Feature Engineering

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
spotify['is_popular'] = (spotify['acousticness_%'] > 60).astype(int)
```

```python
spotify['is_popular'].sum()
```

```python
spotify.info()
```

```python
if spotify['streams'].dtype =='object':
    spotify['streams'] = spotify['streams'].str.replace(',','')
    spotify['streams'] = pd.to_numeric(spotify['streams'],error='coerce')
```

```python



scaler = MinMaxScaler()
spotify['energy_norm'] = scaler.fit_transform(spotify[['energy_%']])
```

```python
spotify['energy_loudness'] = (spotify['liveness_%'] + spotify['energy_norm']) / 2
```

```python
print(spotify[['acousticness_%', 'is_popular', 'energy_%', 'energy_loudness']].head())
```

# Spilitting the Data:

```python
from sklearn.model_selection import train_test_split
```

```python
X = spotify.drop('is_popular', axis=1)
y = spotify['is_popular']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
print(f"Train shape: {X_train.shape}  Test shape: {X_test.shape}")
```

# Applying a machine learning Model :

### Logistic Regression: 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
spotify = spotify.dropna()
```

```python
from sklearn.linear_model import LogisticRegression

logR = LogisticRegression(max_iter=1000)
logR.fit(X_train, y_train)
```

```python
y_pred = logR.predict(X_test)
```

# Model Evalution:

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("\nClassification Report : ", classification_report(y_test, y_pred))
print("\nConfusion Matrix : ",confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
y_probs = logR.predict_proba(X_test)[:,1]
```

```python
#Calcluate FPR,TPR and Thresholds

fpr, tpr, thresholds = roc_curve(y_test, y_probs)
```

```python
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label='Logistic Regression')
plt.plot([0,1],[0,1], linestyle='--', color='gray')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
auc = roc_auc_score(y_test, y_probs)
print(f"AUC score: {auc: .4f}")
```

# Model Imporvement

```python
from sklearn.model_selection import GridSearchCV
```

### GridSearchCv

```python
param_grid = {
    'C' : [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver' : ['liblinear']
}

logR = LogisticRegression(max_iter = 1000)

grid = GridSearchCV(logR, param_grid, cv=5, scoring='roc_auc')
grid.fit(X_train, y_train)

print('Best Parameters:', grid.best_params_)
print('Best AUC Score:', grid.best_score_)
```
